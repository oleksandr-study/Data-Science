```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
# Прочитайте отриману таблицю, використовуючи метод head

jun2017 = pd.read_csv('2017_jun_final.csv')
print(jun2017.head())
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

jun2017.isnull().sum()
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і 
# переконайтеся, що залишився тільки стовпець "Мова.програмування"

new_jun2017 = jun2017.dropna(axis=1, how='any').combine_first(jun2017[["Мова програмування"]])

new_jun2017.isnull().sum()
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
# Визначте новий розмір таблиці за допомогою методу shape

new2_jun2017 = jun2017.dropna(axis=1, how='any')
print(f"Number of rows {new2_jun2017.shape[0]} and columns {new2_jun2017.shape[1]}")
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
# Визначте розмір таблиці python_data за допомогою методу shape

python_data = jun2017.loc[jun2017['Мова програмування'] == 'Python']
print(f"Number of rows {python_data.shape[0]} and columns {python_data.shape[1]}")
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

result = python_data.groupby(by=['Посада'])
result.head()
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", 
# виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне
# та максимальне значення у стовпчику "Зарплата.в.місяць"

new_python_data = python_data.groupby(by=['Посада'])
min_max_python_data = new_python_data.agg({'Зарплата на місяць' : ['min', 'max']})
print(min_max_python_data)
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє
# значення заробітної плати на місяць. Використовуйте її для
# методу apply та створіть новий стовпчик "avg"

def fill_avg_salary(args):
    res = args["Зарплата на місяць"].mean()
    return res

min_max_python_data['avg'] = min_max_python_data.apply(fill_avg_salary, axis=1)
min_max_python_data.head()
```

```python
# Створіть описову статистику за допомогою
# методу describe для нового стовпчика.

min_max_python_data.describe()
```

```python
# Збережіть отриману таблицю в CSV файл
min_max_python_data.to_csv("salary_average.csv",sep=';', index=False, encoding='cp1251')
```
