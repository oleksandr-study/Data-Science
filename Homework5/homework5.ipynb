```python
from os import listdir
from os.path import isfile, join
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
datapath = Path("data")
activities = listdir(datapath)
activities
```

```python
idle = pd.DataFrame()
files = listdir(join(datapath, 'idle'))
for datafile in files:
    file_inf = pd.read_csv(join(join(datapath, 'idle'), datafile))
    idle = pd.concat([idle, file_inf], ignore_index=True)
    
running = pd.DataFrame()
files = listdir(join(datapath, 'running'))
for datafile in files:
    file_inf = pd.read_csv(join(join(datapath, 'running'), datafile))
    running = pd.concat([running, file_inf], ignore_index=True)

stairs = pd.DataFrame()
files = listdir(join(datapath, 'stairs'))
for datafile in files:
    file_inf = pd.read_csv(join(join(datapath, 'stairs'), datafile))
    stairs = pd.concat([stairs, file_inf], ignore_index=True)

walking = pd.DataFrame()
files = listdir(join(datapath, 'walking'))
for datafile in files:
    file_inf = pd.read_csv(join(join(datapath, 'walking'), datafile))
    walking = pd.concat([walking, file_inf], ignore_index=True)

print(idle.head(5))
print(running.head(5))
print(stairs.head(5))
print(walking.head(5))
```

```python
idle['activity'] = 0
running['activity'] = 1
stairs['activity'] = 2
walking['activity'] = 3

print(idle.head(5))
print(running.head(5))
print(stairs.head(5))
print(walking.head(5))
```

```python
main_df = pd.concat([idle, running, stairs, walking], ignore_index=True)
main_df
```

```python
y = main_df.activity
X = main_df[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']]
```

```python
features = pd.DataFrame()
features["mean"] = X.mean(axis=1, numeric_only=True)
features["max"] = X.max(axis=1, numeric_only=True)
features["min"] = X.min(axis=1, numeric_only=True)
features["std"] = X.std(axis=1, numeric_only=True)
features["skew"] = X.skew(axis=1, numeric_only=True)
features["median"] = X.median(axis=1, numeric_only=True)
features
```

```python
X = pd.concat([X, features], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

svm_model = SVC()
svm_model.fit(X_train, y_train)

svm_predictions = svm_model.predict(X_test)

svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_classification_report = classification_report(y_test, svm_predictions)
svm_confusion_matrix = confusion_matrix(y_test, svm_predictions)

print("SVM classifier results:")
print("Accuracy:", svm_accuracy)
print("Classification report:")
print(svm_classification_report)
print("Confusion matrix:")
print(svm_confusion_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_classification_report = classification_report(y_test, rf_predictions)
rf_confusion_matrix = confusion_matrix(y_test, rf_predictions)

print("Random forest classifier results:")
print("Accuracy:", rf_accuracy)
print("Classification report:")
print(rf_classification_report)
print("Confusion matrix:")
print(rf_confusion_matrix)
```
